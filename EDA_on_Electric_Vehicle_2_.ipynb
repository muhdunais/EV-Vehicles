**Load the Dataset and Check Basic Information.**

```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data=pd.read_csv("/content/RS_Session.csv")
```

```python
data = pd.read_csv("/content/RS_Session.csv")
print(data.head())
print(data.info())
```

**1.Which state has the highest and lowest total vehicle sales?**

```python

state_vehicle_sales = data.groupby('State/ UT')['Total Vehicles Sold'].sum()
state_with_highest_sales = state_vehicle_sales.idxmax()
state_with_lowest_sales = state_vehicle_sales.idxmin()
print("State with the highest total vehicle sales:", state_with_highest_sales)
print("State with the lowest total vehicle sales:", state_with_lowest_sales)
```

**2. How does total vehicle sales vary by state over time ?**

```python
import matplotlib.pyplot as plt

total_vehicles = data.groupby("State/ UT")["Total Vehicles Sold"].sum().sort_values(ascending=False)
plt.figure(figsize=(12, 6))
plt.bar(total_vehicles.index, total_vehicles.values, color='skyblue', edgecolor='black')
plt.xticks(rotation=90)
plt.xlabel("State/ UT", fontsize=12)
plt.ylabel("Total Vehicles Sold", fontsize=12)
plt.title("Total Vehicles Sold by State", fontsize=14, fontweight='bold')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

plt.figure(figsize=(12, 6))
plt.scatter(total_vehicles.index, total_vehicles.values, color='skyblue', edgecolor='black')
plt.xticks(rotation=90)
plt.xlabel("State/ UT", fontsize=12)
plt.ylabel("Total Vehicles Sold", fontsize=12)
plt.title("Total Vehicles Sold by State", fontsize=14, fontweight='bold')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()



```

This bar graph represents the  state/UT with the highest sales could be a major urban or economically strong region.
The lower values  represent less populated or less economically developed areas.

The histogram represents the distribution of the percentage share of electric vehicles (EVs) in total vehicle sales. **6. How to check the comparison between vehicles sold in each state :**

```python
state_vehicle_range = data.groupby('State/ UT')['Total Vehicles Sold'].agg(['min', 'max'])
state_vehicle_range = state_vehicle_range.rename(columns={'min': 'Minimum Vehicles Sold', 'max': 'Maximum Vehicles Sold'})
print(state_vehicle_range)
plt.figure(figsize=(12, 6))
plt.bar(state_vehicle_range.index, state_vehicle_range['Minimum Vehicles Sold'], label='Minimum Vehicles Sold')
plt.bar(state_vehicle_range.index, state_vehicle_range['Maximum Vehicles Sold'], label='Maximum Vehicles Sold')
plt.xlabel('State/ UT')
plt.ylabel('Number of Vehicles Sold')
plt.title('Comparison of Minimum and Maximum Vehicles Sold by State')
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

The bar chart compares the minimum and maximum number of vehicles sold across different states and union territories. Each state has two bars: one representing the minimum vehicle sales and the other showing the maximum. This visualization helps identify variations in vehicle sales within each region, highlighting fluctuations over time or across different categories. States with a large difference between their minimum and maximum sales may experience seasonal trends, economic influences, or policy-driven demand shifts. Meanwhile, states with similar minimum and maximum sales indicate a more stable vehicle market. The x-axis labels are rotated for readability, and a legend distinguishes the two categories, making it easier to compare trends across states.

**15. How can we visualize correlations Total ,Total Vehicles sold,% Share of EV in total vehicles sold?**

```python
# Convert columns
numeric_columns = ['Total EV', 'Total Vehicles Sold', '% of Share of EV in Total Vehicles Sold']
for column in numeric_columns:
    data[column] = pd.to_numeric(data[column], errors='coerce')


num_data = data.select_dtypes(include=['number'])
correlation_matrix = num_data.corr()

# Plot heatmap
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", linewidths=0.5)
plt.show()
```

The heatmap visualizes the correlation between key numerical variables related to electric vehicle (EV) sales, including the total number of EVs sold, total vehicles sold, and the percentage share of EVs in total vehicle sales. It helps identify relationships between these factors by displaying correlation values, where a positive correlation (closer to +1) indicates that as one variable increases, the other also tends to increase, while a negative correlation (closer to -1) suggests an inverse relationship. If a strong positive correlation exists between total EV sales and total vehicle sales, it implies that states with higher overall vehicle sales also tend to have higher EV adoption. Conversely, a weak or negative correlation would suggest that EV market penetration does not directly depend on total vehicle sales. The color gradient in the heatmap, ranging from cool blue to warm red, provides a quick visual reference for identifying strong and weak correlations. This analysis offers insights into regional EV adoption trends and helps in understanding whether high vehicle sales translate to greater EV adoption across states.

**18.How can we visualize total vehicles sold using a pie
chart?**

```python
import matplotlib.pyplot as plt

vehicle_counts = data["Total Vehicles Sold"].value_counts()


threshold = 0.02 * vehicle_counts.sum()
filtered_counts = vehicle_counts[vehicle_counts > threshold]
other_count = vehicle_counts[vehicle_counts <= threshold].sum()


if other_count > 0:
    filtered_counts["Others"] = other_count


plt.figure(figsize=(8, 8))


plt.pie(filtered_counts, labels=filtered_counts.index,
        autopct="%1.1f%%", colors=plt.cm.Paired.colors,
        startangle=140, wedgeprops={'edgecolor': 'black'},
        pctdistance=0.85)


centre_circle = plt.Circle((0, 0), 0.70, fc='white')
plt.gca().add_patch(centre_circle)

plt.title("Distribution of Total Vehicles Sold", fontsize=14, fontweight="bold")


plt.show()

```

The pie chart illustrates the distribution of total vehicle sales across different states or regions by showing the proportion of each sales value as a percentage of the total. Each slice represents a specific total vehicle sales figure, with the percentage labels providing insight into how frequently each value appears in the dataset.

However, using a pie chart for this type of data may not be the most effective choice, as it primarily highlights the frequency of specific sales numbers rather than providing a clear comparison between states. If the dataset has many unique sales values, the chart may become cluttered and difficult to interpret. A better alternative could be a bar chart or a histogram, which would more effectively showcase the distribution of vehicle sales across regions.
